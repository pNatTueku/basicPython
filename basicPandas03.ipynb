Working with Pandas and Numpy

```python
# step 1: import pandas and numpy
import pandas as pd
import numpy as np
```

```python
# step 2: set NumPy's random number generation seed to 101
np.random.seed(101)
```

```python
# step 3: create a Numpy matrix of 100 rows by 5 columns consisting of random integer from 1-100
ranData = np.random.randint(1, 100, (100, 5))
ranData
```

```python
# step 4: Now use pd.DataFrame() to read in this numpy array as a dataframe
df = pd.DataFrame(ranData)
df
```

```python
# step 5: rename the columns to be ['C1', 'C2', 'C3', 'C4', 'label']
df.columns = ['C1', 'C2', 'C3', 'C4', 'label']
```

```python
df.head()
```

```python
# step 6: create data frame with the columns ['A', 'B', 'C', 'D'] with each column having 50 rows of random numbers for data. The random numbers should be between 0 and 100.
randomNumbers = np.random.randint(0, 100, (50, 4))
colNames = ['A', 'B', 'C', 'D']
```

```python
df_data = pd.DataFrame(data = randomNumbers, columns = colNames)
df_data
```
