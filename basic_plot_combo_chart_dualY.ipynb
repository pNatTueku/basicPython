```python
def dual_y_demo():

    labels = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])

    x = np.arange(labels.size)
    y1 = np.random.normal(800, 90, labels.size) # random data - normal distribution with mean = 800, sd = 90

    fig, ax1 = plt.subplots()
    ax1.bar(x, y1)
    ax1.set_xticks(x)
    ax1.set_xticklabels(labels)

    ax2 = ax1.twinx() # create twin x
    y2 = np.random.normal(80, 10, labels.size) # random data - normal distribution with mean = 80, sd = 10
    ax2.plot(x, y2, marker = 'o', color = 'green') # share same x

    plt.show()

if __name__ == '__main__':
    dual_y_demo()
```

```python
def dual_y_demo():

    labels = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])

    x = np.arange(labels.size)
    y1 = np.random.normal(800, 90, labels.size) # random data - normal distribution with mean = 800, sd = 90

    fig, ax1 = plt.subplots()
    ax1.bar(x, y1, color = 'purple', alpha = .3)
    ax1.tick_params("y", colors = 'purple')
    ax1.set_ylabel('Sales', color = 'purple', fontsize = 12)

    ax1.set_xticks(x)
    ax1.set_xticklabels(labels)

    ax2 = ax1.twinx() # create twin x
    y2 = np.random.normal(80, 10, labels.size) # random data - normal distribution with mean = 80, sd = 10
    ax2.plot(x, y2, marker = 'o', color = 'green') # share same x
    ax2.tick_params("y", colors = 'green')
    ax2.set_ylabel('Profit', color = 'green', fontsize = 12)

    plt.show()

if __name__ == '__main__':
    dual_y_demo()
```

```python
def dual_y_demo():

    labels = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])

    x = np.arange(labels.size)
    y1 = np.random.normal(800, 90, labels.size) # random data - normal distribution with mean = 800, sd = 90

    fig, ax1 = plt.subplots()
    ax1.bar(x, y1, color = 'purple', alpha = .3)
    ax1.tick_params("y", colors = 'purple')
    ax1.set_ylabel("Sales ('000)", color = 'purple', fontsize = 12)

    ax1.set_xticks(x)
    ax1.set_xticklabels(labels)
    ax1.legend()
    ax1.set_title('Sales vs Profit (Jan-Jun 2023)')

    ax2 = ax1.twinx() # create twin x
    y2 = np.random.normal(80, 10, labels.size) # random data - normal distribution with mean = 80, sd = 10
    ax2.plot(x, y2, marker = 'o', color = 'green') # share same x
    ax2.tick_params("y", colors = 'green')
    ax2.set_ylabel('Profit', color = 'green', fontsize = 12)

    plt.show()

if __name__ == '__main__':
    dual_y_demo()
```

```python
def dual_y_demo():

    labels = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])

    x = np.arange(labels.size)
    y1 = np.random.normal(800, 90, labels.size) # random data - normal distribution with mean = 800, sd = 90

    fig, ax1 = plt.subplots()
    ax1.bar(x, y1, color = 'purple', alpha = .3)
    ax1.tick_params("y", colors = 'purple')
    ax1.set_ylabel("Sales ('000)", color = 'purple', fontsize = 12)

    ax1.set_xticks(x)
    ax1.set_xticklabels(labels)
    ax1.legend()
    ax1.set_title('Sales vs Profit (Jan-Jun 2023)')

    ax2 = ax1.twinx() # create twin x
    y2 = np.random.normal(80, 10, labels.size) # random data - normal distribution with mean = 80, sd = 10
    ax2.plot(x, y2, marker = 'o', color = 'green') # share same x
    ax2.tick_params("y", colors = 'green')
    ax2.set_ylabel('Profit', color = 'green', fontsize = 12)
    ax2.set_ylim(ymin = 0)

    plt.show()

if __name__ == '__main__':
    dual_y_demo()
```
