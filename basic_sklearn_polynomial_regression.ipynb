```python
# get csv datd from Aj. Prasert's Github

url = 'https://github.com/prasertcbs/basic-dataset/raw/master/worldbank_gdp_gini_birth_death_rate.csv'

data = pd.read_csv(url)
data.head(3)
```

```python
# setting

X_col = 'year'
y_col = 'Death rate, crude (per 1,000 people)'
country = 'Thailand'
```

```python
df = data[data['country'] == country].copy()
df
```

```python
sns.lmplot(data = df, x = X_col, y = y_col, fit_reg = True, scatter_kws = {'alpha': 0.2, 'color': 'green'}, line_kws = {'color': 'salmon'});
```

### scikit-learn

```python
# import libraries

import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
print(f'sklearn version: {sk.__version__}')
```

```python
X = df[[X_col]]
y = df[y_col]
```

```python
X.head()
```

```python
y.head()
```

```python
# setting

test_size = 0.2 # 20% for test size
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_size, random_state = 1)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

#### linear regression

simple linear regression model

y = B0 + B1X + E

```python
lm = LinearRegression() # create a model
lm
```

```python
lm.fit(X_train, y_train) # fit/train a model
```

```python
lm.score(X_train, y_train) # get accuracy score: R-squared
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
# visualize data

fig = plt.figure(figsize = (10, 6))
plt.scatter(X, y, color = 'green', alpha = .2)
```

```python
# visualize data

fig = plt.figure(figsize = (10, 6))
plt.scatter(X, y, color = 'green', alpha = .2)
plt.plot(X, lm.predict(X), color = 'red', alpha = .5) #  lm.predict(X) is y_predicted
plt.title('Linear Regression')
plt.xlabel(X_col)
plt.ylabel(y_col)
```

```python
lm.predict([[1990], [2021]])
```
