```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Boxplot
- It is created by the quartiles of dataset, whic devides the number into 4 pieces based on their distribution.
    - **Median**: Value in the middle of the distribution
    - **Lower quartile**: Midpoint between the median and lowest value in the range
    - **Upper quartile**: Midpoint between the median and highest value in the range
    - **Lower boundary**: The lowest value in the distribution
    - **Higher boundary**: The highest value in the distribution

```python
# create a dataframe 

np.random.seed(10)
df = pd.DataFrame(np.random.rand(10,3), columns = ['col1', 'col2', 'col3'])
df.head(3)
```

```python
# plot a single column data

boxplot1 = df.boxplot(column = 'col1')
```

```python
# plot a single column data

boxplot1 = df.boxplot(column = 'col1')
boxplot1.plot()
```

```python
# plot multiple columns

boxplot2 = df.boxplot(column = ['col1', 'col2', 'col3'])
boxplot2.plot()
```

```python
# define color

# plot multiple columns

boxplot2 = df.boxplot(column = ['col2'], color = 'green')
boxplot2.plot()
```

```python
# plot multiple columns with defined color

boxplot2 = plt.boxplot(df, patch_artist = True)
colors = ['orange', 'green', 'blue']

for patch, color in zip(boxplot2['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```

```python
# plot multiple columns with defined color
# add title

boxplot2 = plt.boxplot(df, patch_artist = True)
colors = ['orange', 'green', 'blue']

for patch, color in zip(boxplot2['boxes'], colors):
    patch.set_facecolor(color)

plt.title('Boxplot of Random Numbers')
plt.show()
```

```python
# plot multiple columns with defined color
# add title

boxplot2 = plt.boxplot(df, patch_artist = True)
colors = ['orange', 'green', 'blue']

for patch, color in zip(boxplot2['boxes'], colors):
    patch.set_facecolor(color)

plt.title('Boxplot of Random Numbers', size = 20)
plt.show()
```
