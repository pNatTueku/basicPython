```python
# import libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
def heart():
    # http://mathworld.wolfram.com/HeartCurve.html

    t = np.linspace(0, 2 * np.pi, 360)
    print(t)
    x = 16 * np.sin(t) ** 3
    y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)
    
    plt.plot(x, y)
    plt.plot(x + 10, y)
    plt.plot(x + 20, y)
    plt.show()
```

```python
heart()
```
