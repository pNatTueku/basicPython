```python
fig, ax1 = plt.subplots() # create an empty using plt.subplots() and create the figure (fig) and it axis (ax1)

years = [2020, 2021, 2022, 2023] # x variable
y1 = [1, 2, 1, 2] # 1st y variable
y2 = [10, 13, 15, 9] # 2nd y variable
y3 = [110, 127, 115, 132] # 3rd y variable

ax1.plot(years, y1, color = 'blue') # plot the data start with x then y variables

ax2 = ax1.twinx() # makes ax1 and ax2 using the shared x axis
ax2.plot(years, y2, color = 'green')# plot the second y and change the line color

ax3 = ax1.twinx() # the all 3 y will share the same x
ax3.plot(years, y3, color = 'orange')
# ax3.spines['right'].set_position(('outward', 60)) # put the 3rd y axis to be spide a little bit to the right handside by outward and move to 60
ax3.spines['right'].set_position(('axes', 1.15)) # do the same thing as above ax3

# set the y labels
ax1.set_ylabel('y1', color = 'blue')
ax2.set_ylabel('y2', color = 'green')
ax3.set_ylabel('y3', color = 'orange')

 # change y axis number's color
ax1.tick_params(axis = 'y', colors = 'blue')
ax2.tick_params(axis = 'y', colors = 'green')
ax3.tick_params(axis = 'y', colors = 'orange')

# change color to the actual y
ax2.spines['right'].set_color('green')
ax3.spines['right'].set_color('orange')
ax3.spines['left'].set_color('blue') # now we see the axis color change

plt.show()

# save figure
fig.tight_layout() # to include the 3rd axis
fig.savefig('3_axis_v2.png') # this one we will not see the 3rd axis as we moved it a way
```
