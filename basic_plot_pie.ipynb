```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def pie_demo():

    label = ("China", "Russia", "Japan", "Korea") # label
    val = (1.8, 1, 0.8, 0.5) # values

    plt.pie(val, labels = label)
    plt.show()

# test 
if __name__ == '__main__':
    pie_demo()
```

```python
def pie_demo():

    label = ("China", "Russia", "Japan", "Korea") # label
    val = (1.8, 1, 0.8, 0.5) # values

    plt.pie(val, labels = label, startangle = 90, autopct = "%1.1f%%") # anticlockwise
    plt.axis("equal")
    plt.show()

# test 
if __name__ == '__main__':
    pie_demo()
```

```python
def pie_demo():

    label = ("China", "Russia", "Japan", "Korea") # label
    val = (1.8, 1, 0.8, 0.5) # values

    plt.pie(val, labels = label, startangle = 90, autopct = "%1.2f%%", explode = (0, 0, 0.1, 0)) # anticlockwise, explode Japan (No.3)
    plt.axis("equal")
    plt.show()

# test 
if __name__ == '__main__':
    pie_demo()
```

```python
def pie_demo():

    label = ("China", "Russia", "Japan", "Korea") # label
    val = (1.8, 1, 0.8, 0.5) # values

    plt.pie(val, labels = label, startangle = 90, autopct = "%1.2f%%", explode = (0, 0, 0.5, 0)) # anticlockwise, explode Japan (No.3)
    plt.axis("equal")
    plt.show()

def pie_demo2(): # apply numpy array

    label = np.array(["China", "Russia", "Japan", "Korea"]) # label
    val = (1.8, 1, 0.8, 0.5) # values
    explode = np.zeros(label.size)
    # explode[2] = .1
    explode[np.where(label == "Russia")] = .1
    explode[np.where(label == "Japan")] = .1
    colors = ["red", 'orange', 'pink', 'green']

    plt.pie(val, labels = label, startangle = 90, autopct = "%1.2f%%", explode = explode, colors = colors) # anticlockwise, explode Japan (No.3)
    plt.axis("equal")
    plt.show()

# test 
if __name__ == '__main__':
    pie_demo2()
```
