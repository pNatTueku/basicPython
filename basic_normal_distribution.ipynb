```python
nums = rn_default.normal(size = 10000)
plt.hist(nums, bins = 50)
plt.show()
```

```python
nums.min(), nums.max()
```

```python
bins = 100
bin_width = (nums.max() - nums.min()) / bins

bins, bin_width
```

```python
histogram_area = len(nums) * bin_width
histogram_area
```

```python
# generate data

x = np.linspace(-4, 4, 101)
x
```

```python
len(x)
```

```python
# plot

plt.plot(x, scipy.stats.norm.pdf(x))
plt.show()
```

```python
# compare our histogram of observations with the theoretical curve

x = np.linspace(nums.min(), nums.max(), 101)
x
```

```python
# plot

plt.hist(nums, bins = 100)
plt.plot(x, scipy.stats.norm.pdf(x) * histogram_area)
plt.show()
```
