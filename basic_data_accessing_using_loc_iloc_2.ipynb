```python
# import libraries

import pandas as pd 
import yfinance as yf
from datetime import date
```

```python
name = 'MSFT'
start_date = '2024-02-01'
end_date = date.today()
end_date
```

```python
# load data to a dataframe

df = yf.download(name, start = start_date, end = end_date, progress = False)
df.head()
```
