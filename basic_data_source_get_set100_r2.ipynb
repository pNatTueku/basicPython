```python
set100_symbol = pd.read_html('https://en.wikipedia.org/wiki/SET50_Index_and_SET100_Index')[0]
set100_symbol
```

```python
set100_symbol['Symbol2'] = set100_symbol['Symbol'].astype(str) + '.BK'
set100_symbol
```

```python
set100_symbol['Symbol2'].values
```

```python
import yfinance as yf
```

```python
# get csv file for each Symbol

for symbol in set100_symbol['Symbol2'].values:

    data = yf.download(symbol, progress = True) # download and save each symbol in the list
    data.to_csv(f'Data/SET100_download_r2/{symbol}.csv')
```
