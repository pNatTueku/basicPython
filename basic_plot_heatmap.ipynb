```python
# import libraries

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# create random data using Numpy

rand_data = np.random.rand(5, 5) # 5x5
rand_data
```

```python
# plot

sns.heatmap(rand_data, cmap = 'Reds')
plt.title("Heatmanp of Random Data")
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
```
