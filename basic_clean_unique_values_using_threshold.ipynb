`Check if data in columns has unique value` and `screen columns from DataFrame` if those columns have the `count higher than threshold`

```python
# import libraries

import pandas as pd
import numpy as np
```

```python
# create a dictionary of data

data_dict = {
    'col_a': [np.nan] + [0]*18 + [np.nan],
    'col_b': [1]*20,
    'col_c': [1, 2, 3, 4, 5]*4,
    "col_d": [np.nan]*20,
    "col_e": [0, 1]*10
}
```

```python
# create a DataFrame

df = pd.DataFrame(data = data_dict)
df
```

```python
# create a function to screen unique using threshold

def unique_counts(s, threshold = 0):
    counts = s.value_counts(normalize = True, dropna = False)

    if (counts >= threshold).any():
        return False
    return True
```

```python
# call function and save to variable cols_mask

cols_mask = df.apply(unique_counts, threshold = .95)
cols_mask
```

```python
# clean data using cols_mark

cleaned_df = df.loc[:, cols_mask]
cleaned_df
```
