`Multi-Class Classification` is one of four main types of classification in `Machine Learning` (ML)
- There are > 2 class labels which required encoding to map string values - `No normal or abnormal outcome but examples are classified as belonging to one among a range of known classes`.
- A discrete probability distribution that covers a case where an event will have categorucal outcome. e.g. K in {1, 2, 3, ..., K}
- Examples of multi-class classification are:
    - face classiciation
    - plant species classification
    - character recognition
- Popular algorithms are:
    - K-Nearest Neighbors
    - Decision Tree
    - Naive Bayes
    - Random Forest
    - Gradient Boosting
- A strategy of fitting binary multiple classification model:
    - `One-vs-Rest`: Fit on binary classification model for each class vs all other classes
    - `One-vs-One`: Fit one binary classication model for each pair of classes

```python
# import libraries

import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import make_blobs
from collections import Counter
```

```python
# create a dataset

X, y = make_blobs(n_samples = 500, centers = 3, random_state = 42)
```

```python
# count class label

counter = Counter(y)
counter
```

```python
# get first 10 rows of examples

for i in range(10):
    print(X[i], y[i])
```

```python
# visualize input data (X) and color by output/label (y)

for label, _ in counter.items():
    row_idx = np.where(y == label)[0]

    plt.scatter(X[row_idx, 0], X[row_idx, 1], label = str(label))

plt.legend()
plt.show()
```
