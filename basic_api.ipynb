### currency conversion API (JSON)

```python
url = "https://r2de2-workshop-vmftiryt6q-ts.a.run.app/usd_thb_conversion_rate"
url
```

```python
result_conversion_rate = requests.get(url, auth = ('user', 'pass')) # don't need uth = ('user', 'pass')
result_conversion_rate
```

```python
result_conversion_rate.status_code # if 200 means OK
```

```python
result_conversion_rate.headers["content-type"]
```

```python
result_conversion_rate.encoding
```

```python
result_conversion_rate.text # get data as string format
```

```python
result_conversion_rate.json() # get data as JSON format
```

```python
conversion_rate = result_conversion_rate.json()
conversion_rate
```

```python
type(conversion_rate)
```

```python
assert isinstance(conversion_rate, dict) # return TRUE - No error as it is a dict
```

```python
assert isinstance(conversion_rate, str) # return Error 
```

```python
# convert to Pandas dataframe
import pandas as pd

conversion_rate_df = pd.DataFrame(conversion_rate)
conversion_rate_df.head(4)
```

```python
# transform index to a comlunm date as we will join it with other table

conversion_rate_df = conversion_rate_df.reset_index()
conversion_rate_df
```

```python
conversion_rate_df = conversion_rate_df.rename(columns = {'index' : 'date'})
conversion_rate_df
```

```python
conversion_rate_df[:5]
```

```python
conversion_rate_df.info()
```

```python
# transform timestamp to date

conversion_rate_df['date'] = pd.to_datetime(conversion_rate_df['date']).dt.date
conversion_rate_df
```
