### Merge PDF files

```python
# import a library

import os
from PyPDF2 import PdfMerger
```

```python
x = [a for a in os.listdir() if a.endswith(".pdf")]
x
```

```python
merger = PdfMerger()

for pdf in x:
    merger.append(open(pdf, 'rb'))

with open("result.pdf", "wb") as fout:
    merger.write(fout)
```

### Watermark the pdf

```python
# import libraries

from pathlib import Path
from PyPDF2 import PdfWriter, PdfReader, Transformation
```

```python
# create a function

def stamp(pdf_path: Path, 
          pdf_watermark: Path, 
          pdf_result: Path,
):
    watermark_page = PdfReader(pdf_watermark).pages[0] # take the first page

    writer = PdfWriter()

    reader = PdfReader(pdf_path)

    page_indices = list(range(0, len(reader.pages)))

    for index in page_indices:
        content_page = reader.pages[index]
        # content_page.mergeTransformedPage(watermark_page, Transformation())
        content_page.mergeTransformedPage(watermark_page, Transformation())
        writer.add_page(content_page)

    with open(pdf_result, 'wb') as fp:
        writer.write(fp)
```

```python
stamp(
    Path('files/example.pdf'),
    Path('files/panatda.pdf'),
    Path('files/my_watermarked.pdf')
)
```
