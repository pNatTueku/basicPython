```python
def demo1():
    plt.plot([1, 3, 5, 7, 4])
    plt.show()

def demo2():
    fig = plt.figure()

    fig.add_subplot(121)
    plt.plot([1, 3, 5, 7, 4])

    fig.add_subplot(122)
    plt.bar(range(5), [1, 3, 5, 7, 4])

    plt.show()

if __name__ == '__main__':
    # demo1()
    demo2()
```
