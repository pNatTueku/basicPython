```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# create a dataframe

replaced_part_dict = {
    "Replace screw": 77,
    "Replace shockmount": 49,
    "Replace fan": 22,
    "Replace retainer": 9,
    "Replace gear": 5,
    "Replace LED": 5,
    "Replace nut": 4,
    "Replace gasket": 1
}

replaced_part_df = pd.DataFrame(replaced_part_dict.items(), columns = ['repair_action', 'count'])
replaced_part_df
```

```python
# visualize repair action count and percentage

plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.autolayout'] = True

ax = replaced_part_df.plot.barh(x = 'repair_action', y = 'count')
```

```python
# visualize repair action count and percentage

plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.autolayout'] = True

ax = replaced_part_df.plot.barh(x = 'repair_action', y = 'count')
ax.invert_yaxis()
```

```python
# visualize repair action count and percentage

plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.autolayout'] = True

ax = replaced_part_df.plot.barh(x = 'repair_action', y = 'count')
ax.invert_yaxis()
ax.set(title = 'Repair Action') # add title
sns.despine(bottom = False) # remove right and top axis
```

```python
# visualize repair action count and percentage

plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.autolayout'] = True

ax = replaced_part_df.plot.barh(x = 'repair_action', y = 'count')
ax.invert_yaxis()
ax.set(title = 'Repair Action') # add title
sns.despine(bottom = False) # remove right and top axis

total = sum(replaced_part_df['count']) # sum total

for p in ax.patches:

    count_repair_action = '{:.0f}'.format(p.get_width())
    percent_repair_action = '{:.2f}'.format(100*p.get_width()/total)

    x = p.get_x() + p.get_width() + 0.02
    y = p.get_y() + p.get_height() / 2

plt.show()
```

```python
# visualize repair action count and percentage

plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.autolayout'] = True

ax = replaced_part_df.plot.barh(x = 'repair_action', y = 'count')
ax.invert_yaxis()
ax.set(title = 'Repair Action') # add title
sns.despine(bottom = False) # remove right and top axis

total = sum(replaced_part_df['count']) # sum total

for p in ax.patches:

    count_repair_action = '{:.0f}'.format(p.get_width())
    percent_repair_action = '{:.2f}'.format(100*p.get_width()/total)

    x = p.get_x() + p.get_width() + 0.02
    y = p.get_y() + p.get_height() / 2

    ax.annotate(str(count_repair_action) + ', ' + str(percent_repair_action), (x, y))

plt.show()
```
