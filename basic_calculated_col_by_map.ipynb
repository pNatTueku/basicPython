###  How to create new column with calculated values from column exist

```python
# import a library

import pandas as pd
```

```python
# import csv data to a dataframe

df = pd.read_csv('Iris.csv')
df.sample(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df.Species.value_counts()
```

```python
# create a mapping data by using dictionary

species_num_map = {'Iris-setosa': 0,
                   'Iris-versicolor': 1,
                   'Iris-virginica': 2}
species_num_map
```

```python
# create a new column named "SpeciesNum"

df['SpeciesNum'] = df['Species'].map(species_num_map)
df.head()
```
