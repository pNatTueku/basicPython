```python
# import libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
def qc_plot():

    # setting
    n = 100
    x = np.arange(n)
    d = np.random.normal(150, 5, n)
    m, sd = np.mean(d), np.std(d)
    # plt.hist(d, 30, normed = True)
    plt.plot(x, d, marker = "o", color = "0.7", alpha = 0.7)

    plt.show()

if __name__ == '__main__':
    qc_plot()
```
