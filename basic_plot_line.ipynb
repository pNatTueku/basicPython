```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 3, 7, 2]) # pass list or tuple to plot - y values
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 3, 7, 2]) # pass list or tuple to plot - both x and y value
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 3, 7, 2], "o") # pass list or tuple to plot - both x and y value
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 3, 7, 2], "ro") # pass list or tuple to plot - both x and y value
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 3, 7, 2], 'ro-') # pass list or tuple to plot - both x and y value
plt.show()
```

```python
def demo_plot():
    plt.plot([1, 3, 7, 2], color = 'green', marker = 'o') # no define x, so python will start the x with 0
    plt.plot([1, 5, 3, 4], color = 'blue', linestyle = "--")
    plt.show()

def demo_plot2():
    plt.plot([2014, 2015, 2016, 2017], [1, 3, 7, 2], color = 'green', marker = 'o') # no define x, so python will start the x with 0
    plt.plot(range(2014, 2018), [1, 5, 3, 4], color = 'blue', linestyle = "--")
    plt.show()

def demo_plot3():
    x = range(2017, 2024)
    y = [10, 25, 15, 20, 22, 18, 7]
    plt.plot(x, y)
    plt.show()

def demo_plot4():
    x = range(2017, 2024)
    y = [10, 25, 15, 20, 22, 18, 7]
    plt.plot(x[:4], y[:4], color = 'deepskyblue', label = 'Actual') # slice to plot
    plt.plot(x[3:], y[3:], color = "deepskyblue", linestyle = "--", label = 'Forecast') # slice to plot as the forecast values
    plt.title('Sales Forecast')
    plt.ylabel("Sales ('000)")
    plt.xlabel("Year")
    plt.legend()
    plt.show()

if __name__ == '__main__':
    # demo_plot()
    # demo_plot2()
    # demo_plot3()
    demo_plot4()
```
