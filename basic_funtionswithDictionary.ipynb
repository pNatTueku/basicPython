```python
# import library

import pandas as pd
import numpy as np
```

```python
# create a dataframe

data = {'Name': ['Mark', 'Jay B', 'Jackson', 'Jinyoung', 'Youngjae', 'BamBam', 'Yugyeom'],
'Height': [175, 174, 179, 178, 177, 175, 182],
'Weight': [60, 66, 63, 63, 61, 60, 67]
}

df = pd.DataFrame(data = data)
df
```

```python
df.dtypes
```

```python
df.iloc[:3, :3]
```

```python
# astype function

df = df.astype({'Weight': 'float', 'Height': 'float'})
df
```

```python
df.dtypes
```

```python
# replace function

df['Name'] = df['Name'].replace({'Jay B': 'JB'})
df
```

```python
# add missing data to our dataframe

df.iloc[2:3, 1:3] = np.nan
df
```

```python
# fill na

df = df.fillna({'Weight': 64, 'Height': 179})
df
```

```python
# apply() function

df.apply({'Height': lambda x: x + 0.5, 'Weight': lambda x: x + 1})
```
