### Function

```python
import requests
import pandas as pd

def gov_open_data(api_key, resource_id: str, nrows = 10000):

    """
    >>> gov_open_data("your api-key", "ce7f4a78-71db-4754-9084-edca971903bd", nrows=3)
    """

    # Request headers
    headers = {"api-key": api_key,}

    params = {"resource_id": resource_id, "limit": nrows}
    r = requests.get("https://opend.data.go.th/get-ckan/datastore_search", params, headers = headers)

    j = r.json()
    records = j["result"]["records"]
    return pd.DataFrame(records)
```

```python
gov_open_data(api_key, "d13e3be5-9020-4a78-9a7e-760b2dcbf6be", nrows = 3)
```
