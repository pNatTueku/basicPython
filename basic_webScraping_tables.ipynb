```python
data.items = []

for item in data_list:
    inflation_rate_str = item[-3]
    print(type(inflation_rate_str), inflation_rate_str)
```

```python
data_items = []

for item in data_list:
    year = item[0]
    government_debt_str = item[-4]
    inflation_rate_str = item[-3]
    

    if government_debt_str.endswith('%'):
        government_debt = government_debt_str.replace('%', '')

    if inflation_rate_str.endswith('%'):
        inflation_rate = inflation_rate_str.replace('%', '')

    data_items.append(
        {
        'Year': year,
        'Government Debt (%)': government_debt,
        'Inflation Rate (%)': inflation_rate
        }
    )
```

```python
data_items[0]
```

```python
for item in data_items:
    print(item['Year'], item['Government Debt (%)'], item['Inflation Rate (%)'])
```
