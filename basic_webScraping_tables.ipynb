```python
# read data and prepare to a list of lists data

url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)_per_capita'
tables = pd.read_html(url)
data = tables[1]
data_list = data.values.tolist()[1:]
data_list[0]
```

```python
# create a function to convert data

def check_all_data(imf, wb, un):
    if not imf.isdigit():
        return False
    if not wb.isdigit():
        return False
    if not un.isdigit():
        return False
    return True

data_items = []

for item in data_list:
    country = item[0]
    imf = item[2]
    wb = item[4]
    un = item[6]

    if not check_all_data(imf, wb, un):
        print('skipping', country, imf, wb, un)
        continue

    if country.endswith('\u202f*'): # assume that all countries end with \u202f*
        country = country.replace('\u202f*', '')
    
    data_items.append(
        {
        'country': country,
        'imf': int(imf),
        'wb': int(wb),
        'un': int(un)
        }
    )
    
```

```python
data_items[0]
```

```python
data_items[-1]
```

```python
# process data - calculate the average GDP

processed_data = []

for item in data_items:
    avg = int(round((item['imf'] + item['wb'] + item['un'])/3, 0))

    # append each item
    processed_data.append(
        {
        'country': item['country'],
        'avg gdp pc': avg
        }
    )
```

```python
processed_data[-1]
```

```python
# write data items to csv file

import csv

with open('records_data.csv', 'w') as f:
    csv_writer = csv.DictWriter(f, fieldnames = ['country', 'avg gdp pc'])
    csv_writer.writeheader()
    csv_writer.writerows(processed_data)
```
