`Binary Classification` is one of four main types of classification in `Machine Learning` (ML)
- There are 2 class labels which required encoding to map string values - `normal state: 0, abnormal state: 1`
- Examples of binaru classification are:
    - email spam detection (spam/not spam)
    - churn prediction (churn/not churn)
    - conversion prediction (buy/not buy)
- Popular algorithms are:
    - Logistic
    - K-Nearest Neighbors
    - Decision Tree
    - Support Vector Machine (SVM)
    - Naive Bayes

```python
# import libraries

import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import make_blobs
from collections import Counter
```

```python
# create a dataset

X, y = make_blobs(n_samples = 500, centers = 2, random_state = 42)
```

```python
# count class label

counter = Counter(y)
counter
```

```python
# get first 10 rows of examples

for i in range(10):
    print(X[i], y[i])
```

```python
# visualize input data (X) and color by output/label (y)

for label, _ in counter.items():
    row_idx = np.where(y == label)[0]

    plt.scatter(X[row_idx, 0], X[row_idx, 1], label = str(label))

plt.legend()
plt.show()
```
