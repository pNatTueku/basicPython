### Pandas and Numpy

- create data
- save data
- load data

```python
# step 1: import pandas and numpy

import pandas as pd
import numpy as np
```

```python
# step 2: set NumPy's random number generation seed to 101

np.random.seed(101)
```

```python
# step 3: create a Numpy matrix of 2000 rows by 5 columns consisting of random integer from 1-2000

ranData = np.random.randint(1, 2000, (2000, 5))
ranData
```

```python
# step 4: Now use pd.DataFrame() to read in this numpy array as a dataframe
df = pd.DataFrame(ranData)
df
```

```python
# step 5: rename the columns to be ['C1', 'C2', 'C3', 'C4', 'label']
df.columns = ['C1', 'C2', 'C3', 'C4', 'label']
```

```python
df.head(4)
```

```python
# save data to a csv file

df.to_csv('random_2000_data.csv')
```

```python
# load csv data to a dataframe

def num_of_lines(filename):
    
    with open(filename) as f:
        for row, _ in enumerate(f):
            pass
        return row + 1
```

```python
lines = num_of_lines('random_2000_data.csv')
n = 100

df = pd.read_csv('random_2000_data.csv', skiprows = range(1, lines - n))
df
```
