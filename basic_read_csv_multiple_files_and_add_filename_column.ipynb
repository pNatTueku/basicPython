How to process: `Create a for loop and perform these processes`
- Read in the file
- Create a new column and assign the file name
- create a date column based on the filename
- append the files to dataframe

```python
import pandas as pd
import os
```

```python
# files paths

path = r'multiple_csv/'
files = os.listdir(path)
files
```

```python
combined_df = pd.DataFrame()
combined_df
```

```python
type(combined_df)
```

```python
len(combined_df)
```

```python
for file in files:

    temp_df = pd.read_csv(path + file)
    temp_df['filename'] = file
    temp_df['date'] = file.split('_')[1]
    combined_df = combined_df.append(temp_df)

combined_df
```

```python
all_df = pd.DataFrame()
all_df.shape
```

```python
for file in files:

    temp_df = pd.read_csv(path + file)
    temp_df['filename'] = file
    temp_df['date'] = file.split('_')[1]
    temp_df['ccc_sn'] = file.split('_')[-1][:-4]
    all_df = all_df.append(temp_df)

all_df
```
