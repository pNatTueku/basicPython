```python
# import libraries

from alive_progress import alive_bar
import time
```

```python
items = range(1000)
items
```

```python
len(items)
```

```python
for item in items:
    print(item)
```

```python
# create simple progess bar

with alive_bar(len(items)) as bar:
    for item in items:
        time.sleep(0.05) # sleep time ~50 milliseconds
        bar()
```

```python
outer_items = range(3)
outer_items
```

```python
inner_items = range(5)
inner_items
```

```python
# create nested progess bars

with alive_bar(len(outer_items) * len(inner_items), title = 'Processing') as bar: 
    for _ in outer_items:
        for _ in inner_items: 
            time.sleep(0.1)
            bar()
```

```python
# download simulation with progess bar

from alive_progress import alive_bar
import time

file_size = 1024
chunk_size = 64

with alive_bar(file_size//chunk_size, title = 'Downloading') as bar:

    for _ in range(0, file_size, chunk_size):
        time.sleep(0.5)
        bar()
```

```python
# progress bar with percentage and ETA

from alive_progress import alive_bar
import time

total_items = 50

with alive_bar(total_items, title = 'Processing', spinner = 'dots_waves', bar = 'smooth') as bar:

    for _ in range(total_items):
        time.sleep(0.1)
        bar()
```
