```python
def barv():
    x = range(4) # 0, 1, 2, 3
    y = (20, 15, 5, 7)

    plt.bar(x, y)
    plt.show()

def barh():
    x = range(4) # 0, 1, 2, 3
    y = (20, 15, 5, 7)

    plt.barh(x, y)
    plt.show()

# test
if __name__ == '__main__':
    barv()
    barh()
```

```python
def barv():
    x = range(4) # 0, 1, 2, 3
    y = (20, 15, 5, 7)

    xticks = ("Jackson", "JB", "BamBam", "Mark")
    plt.xticks(x, xticks)
    plt.bar(x, y)
    plt.show()

def barh():
    x = range(4) # 0, 1, 2, 3
    y = (20, 15, 5, 7)

    xticks = ("Jackson", "JB", "BamBam", "Mark")
    plt.yticks(x, xticks) # *** maintain xticks but change from plt.xticks() to plt.yticks()
    plt.barh(x, y)
    plt.show()

# test
if __name__ == '__main__':
    barv()
    barh()
```

```python
def barv():
    x = range(4) # 0, 1, 2, 3
    y = (20, 15, 5, 7)

    xticks = ("Jackson", "JB", "BamBam", "Mark")
    plt.xticks(x, xticks)
    plt.bar(x, y)
    plt.show()

def barh():
    x = range(4) # 0, 1, 2, 3
    y = (20, 15, 5, 7)


    xticks = ("Jackson", "JB", "BamBam", "Mark")
    plt.yticks(x, xticks) # *** maintain xticks but change from plt.xticks() to plt.yticks()
    plt.barh(x, y)
    plt.show()

def bar_subplot():
    x = range(4) # 0, 1, 2, 3
    y = (20, 15, 5, 7)


    xticks = ("Jackson", "JB", "BamBam", "Mark")
    fig, ax = plt.subplots(1, 2) # 1 row and 2 columns
    ax[0].bar(x, y)
    ax[1].barh(x, y)


# test
if __name__ == '__main__':
    barv()
    barh()
    bar_subplot()
```

```python
def barv():
    x = range(4) # 0, 1, 2, 3
    y = (20, 15, 5, 7)

    xticks = ("Jackson", "JB", "BamBam", "Mark")
    plt.xticks(x, xticks)
    plt.bar(x, y)
    plt.show()

def barh():
    x = range(4) # 0, 1, 2, 3
    y = (20, 15, 5, 7)


    xticks = ("Jackson", "JB", "BamBam", "Mark")
    plt.yticks(x, xticks) # *** maintain xticks but change from plt.xticks() to plt.yticks()
    plt.barh(x, y)
    plt.show()

def bar_subplot():
    x = range(4) # 0, 1, 2, 3
    y = (20, 15, 5, 7)


    xticks = ("Jackson", "JB", "BamBam", "Mark")
    fig, ax = plt.subplots(2, 1) # 1 column and 2 rows
    ax[0].bar(x, y, color = 'green') # 1st pic
    ax[1].barh(x, y, color = 'purple') # 2nd pic

    plt.sca(ax[0]) # sca is set current axis
    plt.xticks(x, xticks)

    plt.sca(ax[1])
    plt.yticks(x, xticks)

    fig.tight_layout() # areay of graph
    plt.show()


# test
if __name__ == '__main__':
    # barv()
    # barh()
    bar_subplot()
```
