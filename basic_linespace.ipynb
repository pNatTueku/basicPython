```python
df = bdms.loc['2023-01': '2023-12']
df
```

```python
# simple moving average

def sma(close, window):

    return close.rolling(window = window).mean()
```

```python
# set moving average period = 10 days and call sma function to calculate simple moving average and store to new column
ma_period = 10

df[f'sma_{ma_period}'] = sma(df['Close'], window = ma_period)
df
```

```python
df.iloc[ma_period - 5 : ma_period + 5]
```

```python
df["Close"].plot(figsize = (10, 5));
```

```python
# use linespace

n = np.linspace(10, 40, 4)
n
```

```python
plt.figure(figsize = (10, 5))
plt.plot(df['Close'])

for i in range(0, len(n)):

    plt.plot(sma(df['Close'], window = int(n[i])), label = f'sma_{int(n[i])}', linewidth = 2)

plt.legend()
plt.show()
```
