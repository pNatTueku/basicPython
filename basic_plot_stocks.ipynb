```python
import matplotlib.pyplot as plt
import numpy as np
from pandas_datareader import data  # 0.19 -- it has been changed by yahoo 
# python Pandas Datareader no longer working with Yahoo finance site:stackoverflow.com
# Error will be shown!


def demo():

    s = ["SCC.BK", "PTT.BK", "bbl.bk", "aot.bk", "bh.bk", "kbank.bk", "cpf.bk", "dtac.bk"]

    # plot
    fig, ax = plt.subplots(2, 4)
    print(ax.shape)
    stocks = np.array(s).reshape(ax.shape)
    print(stocks.shape)
    start_dt = "2017-1-1"
    end_dt = "2017-1-31"

    for r in range(ax.shape[0]):
        for c in range(ax.shape[1]):
            df = data.DataReader(stocks[r, c], data_source="yahoo", start=start_dt, end=end_dt)
            # print(df.head())
            a = ax[r, c]
            a.set_title(stocks[r, c])
            a.plot(df["Adj Close"])
            a.tick_params(axis="both", which="major", labelsize=7, color=".7")
            a.spines["left"].set_color(".7")
            a.spines["bottom"].set_color(".7")
            a.spines["top"].set_color("white")
            a.spines["right"].set_color("white")
            
    fig.tight_layout()
    fig.autofmt_xdate()
    plt.show()


if __name__ == '__main__':
    demo()
```

```python
# import libraries
import yfinance as yf 
from pandas_datareader import data as pdr

import matplotlib.pyplot as plt 
import numpy as np 

# create a function and test
def demo():

    s = ["BLA.BK", "PTT.BK", "bbl.bk", "aot.bk", "bh.bk", "kbank.bk", "cpf.bk", "BDMS.BK"] # 1x8

    fig, ax = plt.subplots(2, 4) # 2x4 - 2 rows 4 cols
    print(ax.shape)

    stocks = np.array(s).reshape(ax.shape)
    print(stocks.shape)

    # start and end date
    start_dt = "2022-01-01"
    end_dt = "2022-01-31"

    for r in range(ax.shape[0]):
        for c in range(ax.shape[1]):
            df = pdr.get_data_yahoo(stocks[r, c], start = start_dt, end = end_dt)
            print(df.head())
            a = ax[r, c]
            a.set_title(stocks[r, c])
            a.plot(df["Adj Close"])

            a.tick_params(axis = "both", which = "major", labelsize = 7, color = ".7")
            a.spines["left"].set_color(".7") # reduce axes color
            a.spines["bottom"].set_color(".7")
            a.spines["top"].set_color("white") # set axes color to white
            a.spines["right"].set_color("white")

    fig.tight_layout()
    fig.autofmt_xdate() 
    plt.show()


if __name__ == '__main__':
    demo()
```

```python
ptt = pdr.get_data_yahoo("PTT.BK", start = "2024-01-01", end = "2024-01-31")
ptt.shape
```

```python
# rolling data

look_back = 5 # look back 5 days

ptt['highest'] = ptt['Close'].rolling(look_back).max()
ptt
```

```python
# plt Close and highest

ptt[['Close', 'highest']].plot(figsize = (10, 5))
plt.legend()
plt.show()
```
