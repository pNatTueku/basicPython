```python
# Resampling - Change data frequency such as day to month using average, std, etc.

df.index
```

```python
# change object to datetime

pd.to_datetime(df.index)
```

```python
df.index
```

```python
# change object to datetime then set it as index, require inplace = True as it should be saved

df.set_index(pd.to_datetime(df.index), inplace = True)
```

```python
df.index
```

```python
# Resampling from day to year end frequency

df.resample(rule = 'A').mean() # aplly mean()
```

```python
df.resample(rule = 'A').sum() # aplly sum()
```

```python
df_pct.set_index(pd.to_datetime(df_pct.index), inplace = True)
df_pct.index
```

```python
df_pct.resample(rule = 'A').sum() # apply sum()
```

```python
df_pct.resample(rule = 'M').sum() # month end frequency
```

```python
df_pct.resample(rule = 'W').sum() # week end frequency
```

```python
df_pct.resample(rule = 'W').std() # week end frequency by applying std()
```

```python
df_pct.resample(rule = 'A').std() # week end frequency by applying std()
```

```python
plt.rcParams['figure.figsize'] = (20.0, 8.0)

df_yearly = df_pct['Close'].resample(rule = 'A').sum() # resampling as A -year end frequency
df_yearly.plot(kind = 'bar')
plt.show()
```
