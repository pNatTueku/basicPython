```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def demo():

    labels = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])

    x = np.arange(labels.size)
    y1 = np.random.normal(80, 10, x.size) # normal distribution with mean = 80, sd = 10

    plt.bar(x, y1)
    plt.show()

if __name__ == '__main__':
    demo()
```

```python
def demo():

    labels = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])

    x = np.arange(labels.size)
    y1 = np.random.normal(80, 10, x.size) # normal distribution with mean = 80, sd = 10
    y2 = np.random.normal(95, 12, x.size) # normal distribution with mean = 95, sd = 12

    plt.bar(x, y1)
    plt.plot(x, y2)
    plt.xticks(x, labels)
    plt.show()

if __name__ == '__main__':
    demo()
```

```python
def demo():

    labels = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])

    x = np.arange(labels.size)
    y1 = np.random.normal(80, 10, x.size) # normal distribution with mean = 80, sd = 10
    y2 = np.random.normal(95, 12, x.size) # normal distribution with mean = 95, sd = 12

    plt.bar(x, y1, color = 'deepskyblue', alpha = .3, label = 'we') # alpha for stansparant
    plt.plot(x, y2, color = 'orange', label = 'industry', marker = 'o')
    plt.xticks(x, labels)
    plt.legend()
    plt.title("Sales (Jan-Jun 2023)")
    plt.ylabel("Sales ('000)")
    plt.show()

if __name__ == '__main__':
    demo()
```
