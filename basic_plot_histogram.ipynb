### Histogram

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# normal distribution

mu = 0 # mean
sigma = 10
```

```python
x = mu + sigma * np.random.randn(1000)
x
```

```python
# plot to see distribution

plt.hist(x)
plt.show()
```

```python
plt.style.available
```

```python
# plot to see distribution

plt.style.use('classic')

plt.hist(x)
plt.axvline(mu, linewidth = 2, c = 'red', linestyle = '--')
plt.xlabel('Value of X')
plt.ylabel('Density') # here it is normal distribution
plt.title('Distribution of X', fontsize = 16)
plt.show()
```

```python
# plot to see distribution

plt.style.use('classic')

plt.hist(x, color = 'salmon', bins = 100)
plt.axvline(mu, linewidth = 5, c = 'blue', linestyle = '--')
plt.xlabel('Value of X')
plt.ylabel('Density') # here it is normal distribution
plt.title('Distribution of X', fontsize = 16)
plt.show()
```

```python
# plot to see distribution

plt.style.use('classic')

plt.hist(x, color = 'green', bins = 50)
plt.axvline(mu, linewidth = 3, c = 'blue', linestyle = '--')
plt.xlabel('Value of X')
plt.ylabel('Density') # here it is normal distribution
plt.text(5, 5, f'mean = {mu}, sigma = {sigma}', color = 'blue')
plt.title('Distribution of X', fontsize = 16)
plt.show()
```
