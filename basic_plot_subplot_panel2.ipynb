```python
# import libraries

import matplotlib.pyplot as plt
import numpy as np
```

```python
# create a function and test

def demo():

    fig, ax = plt.subplots(2, 1) # row = 2, column = 1
    print(ax)
    print(type(ax))
    print(ax.shape) # it is 1 dimensional

    ax[0].plot([1, 3, 5, 7, 4])
    ax[1].bar(range(5), [1, 3, 5, 7, 4])

    plt.show()

if __name__ == '__main__':
    demo()
```

```python
# create a function and test

def demo():

    fig, ax = plt.subplots(2, 1)
    # print(ax)
    # print(type(ax))
    # print(ax.shape)

    ax[0].plot([1, 3, 5, 7, 4])
    ax[1].bar(range(5), [1, 3, 5, 7, 4])

    plt.show()

def demo2():

    fig, ax = plt.subplots(2, 3)
    print(ax)
    print(type(ax))
    print(ax.shape) # check dimension - it is 2 dimensional

    for r in range(ax.shape[0]):
        for c in range(ax.shape[1]):

            ax[r, c].plot(np.random.randint(1, 11, 5))
            ax[r, c].set_title("ax[{}, {}]".format(r, c))
    
    plt.show()

if __name__ == '__main__':
    # demo()
    demo2()
```

```python
# create a function and test

def demo():

    fig, ax = plt.subplots(2, 1)
    # print(ax)
    # print(type(ax))
    # print(ax.shape)

    ax[0].plot([1, 3, 5, 7, 4])
    ax[1].bar(range(5), [1, 3, 5, 7, 4])

    plt.show()

def demo2():

    fig, ax = plt.subplots(2, 3, sharex = True, sharey = True) # share x and y
    print(ax)
    print(type(ax))
    print(ax.shape) # check dimension - it is 2 dimensional

    for r in range(ax.shape[0]):
        for c in range(ax.shape[1]):

            ax[r, c].plot(np.random.randint(1, 11, 5))
            ax[r, c].set_title("ax[{}, {}]".format(r, c))
    
    plt.show()

if __name__ == '__main__':
    # demo()
    demo2()
```
