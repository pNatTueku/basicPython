Distribution Plots, or Distplots
- typically layer three plots on top of one another

```python
# import libraries
import numpy as np
import plotly.figure_factory as pff # a special section of ploly that does more complex like a distribution plot 
import plotly.offline as plo
```

```python
# create data with multiple group
snodrass = [.209, .205, .196, .210, .202, .207, .224, .223, .220, .201]
twain = [.225, .262, .217, .240, .230, .229, .235, .217] 
```

```python
# add data to a list of dataset
dataHist = [snodrass, twain]
dataHist
```

```python
# create labels
groupLabels = ['Snodrass Writing', 'Mark Twain Writing']
```

```python
# create distribution plot
fig = pff.create_distplot(dataHist, groupLabels, bin_size = [.005, .005])
fig
```
