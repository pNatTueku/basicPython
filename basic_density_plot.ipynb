```python
# import libraries

import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# create normal distribution data

data = np.random.normal(size = 10000)
data
```

```python
len(data)
```

```python
sns.kdeplot(data = data, fill = True, color = 'salmon')
```

```python
sns.kdeplot(data = data, fill = True, color = 'salmon')
plt.title("Density Plot")
plt.xlabel('Data')
plt.ylabel('Density')
plt.show()
```
