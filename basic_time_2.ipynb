```python
# import function and libraries

import time
import numpy as np

from math import factorial
```

```python
# set time

def main():

    x = np.random.randint(1, 10) # random number when it run
    print("A nmber is " + str(x)) # show number
    print("Factorial of " + str(x) + " is " + str(factorial(x))) # use that number to find a factorial
```

```python
# call main function

main()
```

```python
# use time function to continute working 1 minute

time_begin = time.time() # system time now
time_end = time.time() + 60 * 0.5 # 0.5 minute * 60 seconds = 30 minutes

while time.time() <= time_end:
    main() # call main function 
    time.sleep(5) # add sleep/delay every 5 minutes
```
