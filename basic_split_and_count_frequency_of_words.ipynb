```python
# create a list of delimiters

delimiters = ["\t", "\n", ",", "'", '"', "(", ")", "?"]
delimiters
```

```python
for delimiter in delimiters:
    genre = " ".join(genre.split(delimiter))
genre
```

```python
len(genre)
```

```python
words_in_genre = genre.split()
words_in_genre
```

```python
len(words_in_genre)
```

```python
Counter(words_in_genre).most_common()
```

```python
type(Counter(words_in_genre).most_common())
```

```python
# list of words and it's frequency

list_of_words = Counter(words_in_genre).most_common()
list_of_words
```

```python
# store them to a dataframe

words_and_frequency = pd.DataFrame(list_of_words, columns = ['word', 'frequency'])
words_and_frequency.shape
```

```python
words_and_frequency.value_counts()
```

```python
words_and_frequency.sum()
```

```python
words_and_frequency['frequency'].sum()
```

```python
# save to a csv file

words_and_frequency.to_csv("genre_words_and_freq.csv")
```

### Apply method guided by ChatGPT

```python
# use guideline from ChatGPT

def count_words(sentence):

    # split the sentence into words
    words = sentence.split()

    # display the total number of words in the sentence
    words = re.findall(r'\b\w+\b', sentence) # find all delimiters and count them
    total_words = len(words)

    print(f"Total number of words: {total_words}")
```

```python
count_words(genre) # call a function to count number of words -> used delimiters = ["\t", "\n", ",", "'", '"', "(", ")", "?", "-"]
```
