```python
# import libraries
import requests
import bs4 # beautiful soup
```

```python
url = base_page_url + 'page-1.html'
url
```

```python
response = requests.get(url)
response
```

```python
soup = bs4.BeautifulSoup(response.text, 'lxml')
soup
```

```python
products = soup.select('.product_pod')
len(products)
```

```python
# As there are 50 pages, we are going to check for all pages

five_star_ratings = []

base_page_url = 'http://books.toscrape.com/catalogue/'

for i in range(1, 51):
    # url = base_page_url + 'page-1.html' # this is for page #1
    url = base_page_url + 'page-' + str(i) + '.html' # for all pages in the range

    # get the response
    response = requests.get(url)

    # get beautiful soup
    soup = bs4.BeautifulSoup(response.text, 'lxml')

    # get books by getting product pod
    books = soup.select('.product_pod')

    for book in books:
        # check
        if len(book.select('.star-rating.Five')) > 0:
            five_star_ratings.append((book.select('a')[1]['title']))
```

```python
for title in five_star_ratings:
    print(title)
```
