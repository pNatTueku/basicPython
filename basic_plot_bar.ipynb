```python
def demo1():
    x = range(4) # 0, 1, 2, 3
    y = (20, 25, 40, 30)

    xtick = ('Jackson', 'BamBam', 'Mark', 'JB')
    plt.bar(x, y) # pass x and y to plot
    plt.xticks(x, xtick)
    plt.title("Order by Member \n Dec 2023", color = 'darkgreen', fontsize = 18)
    plt.ylabel(" # Scores")
    plt.show()

if __name__ == '__main__':
    demo1() # call demo1 function
```

```python
def demo1():
    x = range(4) # 0, 1, 2, 3
    y = (20, 25, 40, 30)
    avg = sum(y)/len(x)
    # print(avg)


    xtick = ('Jackson', 'BamBam', 'Mark', 'JB')
    plt.bar(x, y) # pass x and y to plot
    plt.xticks(x, xtick)
    plt.axhline(avg)
    plt.title("Order by Member \n Dec 2023", color = 'darkgreen', fontsize = 18)
    plt.ylabel(" # Scores")
    plt.show()

if __name__ == '__main__':
    demo1() # call demo1 function
```

```python
def demo1():
    x = range(4) # 0, 1, 2, 3
    y = (20, 25, 40, 30)
    avg = sum(y)/len(x)
    # print(avg)


    xtick = ('Jackson', 'BamBam', 'Mark', 'JB')
    plt.bar(x, y, color = 'green') # pass x and y to plot
    plt.xticks(x, xtick)
    plt.axhline(avg, color = 'red', linestyle = '--')
    plt.title("Order by Member \n Dec 2023", color = 'darkgreen', fontsize = 18)
    plt.ylabel(" # Scores", color = 'darkgreen')
    plt.show()

if __name__ == '__main__':
    demo1() # call demo1 function
```
